## Creating Your SparkSession

The SparkSession is the entry point to all Spark functionality. It's like opening your broker's office for the day—everything starts here!

**Why SparkSession?** 
- It's the unified entry point introduced in Spark 2.0
- Replaces the old SparkContext, SQLContext, and HiveContext
- Manages connections to Spark clusters
- Handles configuration settings

📚 [Official Documentation](https://spark.apache.org/docs/latest/api/python/reference/pyspark.sql/api/pyspark.sql.SparkSession.html)


```python
from pyspark.sql import SparkSession

# Create a SparkSession
# Think of this as opening your real estate broker's office
spark = SparkSession.builder \
    .appName("RealEstateAnalytics") \
    .master("local[*]") \
    .getOrCreate()

# Let's see what we created
print("✅ SparkSession created successfully!")
print(f"📊 Spark Version: {spark.version}")
print(f"🏢 App Name: {spark.sparkContext.appName}")
print(f"💻 Master: {spark.sparkContext.master}")
```

## Understanding the Spark Architecture

Let's explore what's happening under the hood. The Spark architecture consists of:

1. **Driver Program**: The "broker" that coordinates everything
2. **Cluster Manager**: Manages resources (like a property management company)
3. **Executors**: The "agents" that process data in parallel
4. **Tasks**: Individual units of work (like showing individual properties)


```python
# Get information about the Spark Context
sc = spark.sparkContext

print("🏗️ Spark Architecture Information:")
print(f"   Driver: {sc.appName}")
print(f"   Master URL: {sc.master}")
print(f"   Default Parallelism: {sc.defaultParallelism}")
print(f"   Spark UI: http://localhost:4040 (if running locally)")

# In a real cluster, you'd see multiple executors
# For local mode, we're using all available CPU cores
print(f"\n💡 Tip: In local mode, Spark uses {sc.defaultParallelism} partitions by default")
```

## Creating Your First DataFrame

Let's create a simple DataFrame to test our setup. Think of this as your first property listing!

**What is a DataFrame?**
- A distributed collection of data organized into named columns
- Similar to a table in a relational database or a pandas DataFrame
- But distributed across multiple machines for parallel processing


```python
# Create a simple DataFrame with property data
# This is like creating your first property listing card

properties_data = [
    ("PROP-0001", "123 Main St", "Springfield", 250000, 3, 2),
    ("PROP-0002", "456 Oak Ave", "Riverside", 350000, 4, 3),
    ("PROP-0003", "789 Park Blvd", "Oakwood", 450000, 5, 4),
]

columns = ["property_id", "address", "city", "price", "bedrooms", "bathrooms"]

# Create DataFrame using DataFrame API
df = spark.createDataFrame(properties_data, schema=columns)

print("🏠 Your First Property DataFrame:")
df.show()
```

```python
# Let's see the schema (structure) of our DataFrame
print("📋 DataFrame Schema:")
df.printSchema()

print("\n📊 DataFrame Information:")
print(f"   Number of rows: {df.count()}")
print(f"   Number of columns: {len(df.columns)}")
```

## Spark Execution Model: Lazy Evaluation

**Important Concept!** Spark uses **lazy evaluation**. This means:

- Transformations (like `select`, `filter`) are **not executed immediately**
- They build up a **logical plan** (like a blueprint)
- Only when you call an **action** (like `show()`, `count()`) does Spark actually execute

**Why?** This allows Spark to optimize the entire query plan before execution—like a broker planning the most efficient route to show multiple properties!

**Transformations** (lazy): `select()`, `filter()`, `groupBy()`, `join()`, etc.
**Actions** (eager): `show()`, `count()`, `collect()`, `write()`, etc.


```python
# Example of lazy evaluation
# This doesn't execute anything yet - just builds a plan
filtered_df = df.filter(df.price > 300000)

print("🔍 This is just a logical plan, not executed yet!")
print(f"   Type: {type(filtered_df)}")
print(f"   This is a DataFrame, but the filter hasn't run yet!")

# Now let's trigger execution with an action
print("\n⚡ Now we trigger execution with show():")
filtered_df.show()
```

## Viewing the Execution Plan

Spark creates an optimized execution plan. Let's see what Spark is planning to do!

**Understanding the Plan:**
- Shows how Spark will execute your query
- Helps you understand optimization opportunities
- Useful for debugging performance issues


```python
# View the logical plan (what we want to do)
print("📐 Logical Plan (What we want):")
filtered_df.explain(extended=False)

print("\n" + "="*60)
print("🔧 Physical Plan (How Spark will do it):")
filtered_df.explain(extended=True)
```
